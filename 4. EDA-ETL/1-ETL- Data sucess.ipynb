```python
# Define the function to extract numeric values
def extraer_valor_numerico(value):
    try:
        # Split the value by whitespace and extract the first part
        numeric_value = value.split()[0]
        # Convert the extracted part to float
        numeric_value = float(numeric_value)
        return numeric_value
    except:
        # If any error occurs during conversion, return None
        return None

# Apply the function to extract numeric values from specific columns
electric_cardata['Range (mile)'] = electric_cardata['Range (mile)'].apply(extraer_valor_numerico)
electric_cardata['Efficiency (Wh/mile)'] = electric_cardata['Efficiency (Wh/mile)'].apply(extraer_valor_numerico)
electric_cardata['FastCharge (mile/h)'] = electric_cardata['FastCharge (mile/h)'].apply(extraer_valor_numerico)
```

```python
# Define the function to convert kilometers to miles
def convert_km_to_mile(km):
    # Conversion factor from kilometers to miles
    conversion_factor = 0.621371
    # Convert kilometers to miles
    mile = km * conversion_factor
    return mile

# Aplicamos la funcion de `convert_km_to_mile` a las columnas 'Range', 'Efficiency' y 'FastCharge'
electric_cardata['Range (mile)'] = electric_cardata['Range (mile)'].apply(convert_km_to_mile)
electric_cardata['Efficiency (Wh/mile)'] = electric_cardata['Efficiency (Wh/mile)'].apply(convert_km_to_mile)
electric_cardata['FastCharge (mile/h)'] = electric_cardata['FastCharge (mile/h)'].apply(convert_km_to_mile)
```

## ETL: `taxi_zones.dbf`

```python
# Leemos el dataset 
taxi_zones = dbfread.DBF(r'C:\Users\melan\Desktop\PG 1\ETL -EDA\Data\1. Originales\taxi_zones.dbf')
taxi_zones = pd.DataFrame(iter(taxi_zones))
taxi_zones
```

```python
# Eliminamos las columnas que no nos interesan
eliminar = ['OBJECTID', 'Shape_Leng', 'Shape_Area']
taxi_zones.drop(eliminar, axis=1, inplace=True)
taxi_zones.head(2)
```

```python
# Renombramos las columnas
taxi_zones = taxi_zones.copy()
taxi_zones.rename(columns={'zone':'Zone', 'borough':'Borough'}, inplace=True)

taxi_zones.head(2)
```

```python
# Se Define the function to map borough names to IDs
def borough_id(borough_name):
    # Dictionary mapping borough names to IDs
    borough_mapping = {
        'Manhattan': 1,
        'Bronx': 2,
        'Brooklyn': 3,
        'Queens': 4,
        'Staten Island': 5
    }
    # Return the ID corresponding to the borough name
    return borough_mapping.get(borough_name, None)
```

```python
# Se implementa la funcion 'borough_id' a la columna 'Borough' para crear la nueva columna 'BoroughID'
taxi_zones['BoroughID'] = taxi_zones['Borough'].apply(borough_id)
taxi_zones.head(2)
```

```python
# Siguiendo con la lógica implementada, asignamos el valor 6 al BoroughID de valor 0 que corresponde a EWR
taxi_zones.loc[taxi_zones['BoroughID'] == 0, 'BoroughID'] = 6
taxi_zones.head(2)
```

```python
def split_date_time(date_time_str):
    # Suponiendo que la cadena de fecha y hora tiene un formato como 'yyyy-mm-dd HH:MM:SS'
    date_time_parts = date_time_str.split(' ')  # Divide la cadena en partes utilizando un espacio como delimitador
    if len(date_time_parts) == 2:
        date_str, time_str = date_time_parts
        return pd.Series({'Date': date_str, 'Time': time_str})
    else:
        # Manejar el caso de datos incompletos o formato incorrecto
        return pd.Series({'Date': None, 'Time': None})  # Puedes ajustar esto según sea necesario
```
